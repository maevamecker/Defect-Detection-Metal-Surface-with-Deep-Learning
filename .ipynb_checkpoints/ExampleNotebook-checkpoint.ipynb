```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import torch
from torch import nn, optim
import os
from torchvision import datasets, transforms, models
from skimage import io, transform
import matplotlib.pyplot as plt
from time import time
```

### This notebook presents an use case of applying neural networks to classify defects on

### Context and possible business use case

In a huge number of industries the process of quality check for final product is not always reliable and prone to error when it is done by human.
For this reason, we can use machine learning algorithms to detect and classify a defect on a surface.
This system exists already in the automotive and aeronautic industry.
Here we will use the algorithms to detect defaut on metallic surface. This model associated to a visual sensor could help to improve quality on the production line.

This notebook presents a case study that aims to classify defects on metal surfaces.
To do this we have a dataset of 1800 images of steel surfaces. There are 6 possible defects, so we have 300 images of steel surfaces for each defect.
The defect will be classify automaticaly by the algorithms to one of the folowing category:
- 0 which corresponds to "Crazing"
- 1 which corresponds to "Inclusions"
- 2 which corresponds to "Patches"
- 3 which corresponds to"Pitted Surface"
- 4 which corresponds to "Rolled in Scale"
- 5 which corresponds to "Scratches"

We started with a sub-dataset of only 90 images: 15 for each defect, dividing the set into two parts:
- training set: 60 images (10 for each defect)
- test set: 30 images (5 for each defect)


###  <span style="color:green"> Step 1: Transformations to apply to each image : resize and transformation to tensor </span>

```python
train_trainsforms = transforms.Compose([transforms.Resize((200,200)),
                                            transforms.ToTensor(),
                                            transforms.Normalize((0.5,), (0.5,)),])
    
test_trainsforms = transforms.Compose([transforms.Resize((200,200)),
                                           transforms.ToTensor(),
                                           transforms.Normalize((0.5,), (0.5,)),])

val_trainsforms = transforms.Compose([transforms.Resize((200,200)),
                                           transforms.ToTensor(),
                                           transforms.Normalize((0.5,), (0.5,)),]) #To evaluate our model later on
```

###  <span style="color:green"> Step 2: Charge data train and test </span>

```python
train_data = datasets.ImageFolder('Data_train/',transform=train_trainsforms)
test_data = datasets.ImageFolder('Data_test/',transform=test_trainsforms)
```

```python
num_train = len(train_data)                               #Number of training sets
indices = list(range(num_train))   
```

```python
num_test = len(test_data)                               #Number of testing sets
indices = list(range(num_test))
```

```python
#the loader with load each time 32 images for one training itteration
#We will set a seed to have the same random batch each time we run the script 
torch.manual_seed(6)
train_loader =  torch.utils.data.DataLoader(train_data,batch_size=32,shuffle=True)
test_loader =  torch.utils.data.DataLoader(test_data,batch_size=32,shuffle=True)
```

```python
#print the size of the data
#the first line prints the size of the images tensor : There are 60 images in a batch and each image has 200 x 2000 pixels
#as we have RGB images the first dimension of the image is 3 and not 1

dataiter = iter(train_loader)
images, labels = dataiter.next()
images = images[:,0,:,:]
print(images.shape)
print(labels.shape)

```

###  <span style="color:green"> Step 3: Detection process </span>

The input size be 40000 neurons (200x200). However, since our images are considered in colour, they have to pass through 3 filters (RGB) for this reason the input layer is composed of 120,000 neurons (40000x3).

We also choose the size for our two hidden layers : 1st layer has 128 neurons and 60 is the number of neurons for the 2nd layer 

The ouput size is equal to 6 because we have 6 class of defects

```python
#the input size be 40000 neurons (200x200). 
#However, since our images are considered in colour, they have to pass through 3 filters (RGB)
#for this reason the input layer is composed of 120,000 neurons (40000x3)
input_size = 120000

#choosen sizes for hidden layer
hidden_sizes = [128,64]

output_size = 6

#We build our network with its different layers

model = nn.Sequential(nn.Linear(input_size, hidden_sizes[0]), #1st layer is linear 
                      nn.ReLU(), #ReLu function between layers to filter negatives files and speed up the process
                      nn.Linear(hidden_sizes[0], hidden_sizes[1]), #2nd layer is linear
                      nn.ReLU(), #ReLu function between layers to filter negatives files and speed up the process
                      nn.Linear(hidden_sizes[1], output_size), #3rd layer is linear 
                      nn.LogSoftmax(dim=1)) #output layer with a LogSoftmax
print(model)
```

```python
print(images.shape)
```

We choose a negative log likelihood loss

```python
criterion = nn.NLLLoss()
images, labels = next(iter(train_loader))
```

```python
images = images.view(images.size(0),-1)
```

```python
logps = model(images) #log probabilities, the output of our neural network model
loss = criterion(logps, labels) #calculate the NLL loss
```

The image tensor consists of 32 images (batch size) and therefore 32 rows and 120000 columns number of pixels in each image multiplied by 3 because of the 3 RGB filters.

```python
print(images.shape)
```

We now apply backward propagation to improve the weights and lower the cost function.

```python
print('Before backward pass: \n', model[0].weight.grad)

loss.backward()

print('After backward pass: \n', model[0].weight.grad)
```

We will use the stochastic gradient descent optimiser.

The Gradient Descent is an iterative process.

We try to reach a good accuracy score by making 30 epochs

```python
torch.manual_seed(6) #We use the seed again to have always the same results 
#the Stochastic Gradient Descent optimizer
#momentum is a parameter that will avoid us getting stuck into a local minimum : a value between 0 and 1
optimizer = optim.SGD(model.parameters(), lr=0.003, momentum=0.9)
time0 = time()

#number of feedforward and backpropagation iterations that we will do to train the network
epochs = 30
for e in range(epochs):
    #we start with an error or loss of zero
    running_loss = 0
    
    #we itterate through the images dataset
    for images, labels in train_loader:
        
        # Flatten images into a long vector
        images = images.view(images.size(0),-1)
    
        # Training pass
        optimizer.zero_grad()
        
        output = model(images)
        loss = criterion(output, labels)
        
        #This is where the model learns by backpropagating
        loss.backward()
        
        #And optimizes its weights here
        optimizer.step()
        
        running_loss += loss.item()
    else:
        print("Epoch {} - Training loss: {}".format(e, running_loss/len(train_loader)))
print("\nTraining Time (in minutes) =",(time()-time0)/60)
```

```python
for images, labels in train_loader:
    print(images.shape)
```

```python
torch.manual_seed(6)
correct_count, all_count = 0, 0
for images,labels in test_loader:
  for i in range(len(labels)):
    img = images[i].view(1, 120000)
    #torch.no_grad(): Context-manager that disabled gradient calculation
    with torch.no_grad():
        logps = model(img)

    #get the scores obtained for each class. 
    ps = torch.exp(logps)
    probab = list(ps.numpy()[0])
    
    #choose the one having the maximum score and compare it to the real label
    pred_label = probab.index(max(probab))
    true_label = labels.numpy()[i]
    
    if(true_label == pred_label):
      correct_count += 1
    all_count += 1

print("Number Of Images Tested =", all_count)
print("\nNumber of images predicted correctly =", correct_count)
print("\nModel Accuracy =", (correct_count/all_count))
```

The testing sample contains 30 images

Our model accuracy is about 0.57 which is not awesome, only 57% of the images were predicted correctly (17 out of 30)

Let's try to get a better score by increasing the number of images

Indeed we have tried to build our model on a train set of 60 images and a test set of 30 images.

Let's try to improve it by building our model on a train set of 1620 images and a test set of 180 images.

We will also tune again our parameters.

###  <span style="color:green"> Step 4: Improving our model</span>

```python
train_data = datasets.ImageFolder('New_data_train/',transform=train_trainsforms) #New data train of 1620 images
test_data = datasets.ImageFolder('New_data_test/',transform=test_trainsforms) # New data test of 180 images
val_data = datasets.ImageFolder('Val_set/',transform=val_trainsforms) #Evaluation set of 3 images to test our model on images it has never seen before
```

```python
num_train = len(train_data)                               #Number of training sets
indices = list(range(num_train))   
```

```python
num_test = len(test_data)                               #Number of testing sets
indices = list(range(num_test))
```

```python
num_val = len(val_data)                               #Number of val sets
indices = list(range(num_val))
```

```python
#the loader with load each time 32 images for one training itteration
torch.manual_seed(12)
train_loader =  torch.utils.data.DataLoader(train_data,batch_size=32,shuffle=True)
test_loader =  torch.utils.data.DataLoader(test_data,batch_size=32,shuffle=True)
val_loader =  torch.utils.data.DataLoader(val_data,batch_size=32,shuffle=True)
```

```python
#print the size of the data
#the first line prints the size of the images tensor : There are 60 images in a batch and each image has 200 x 2000 pixels
#as we have RGB images the first dimension of the image is 3 and not 1

dataiter = iter(train_loader)
images, labels = dataiter.next()
images = images[:,0,:,:]
print(images.shape)
print(labels.shape)
```

```python
#we have 6 classes corresponding to 6 types of defects
#the loader will code these classes from 0 to 5
print(labels)
```

```python
#depth, width, height
images[1].numpy().shape
```

We will now build our neural network with 5 hidden layers with size of respectively: 512,256,128 and 64 neurons 

We increase the number of hidden layers by 2 in order to improve our model accuracy, we also add a regularisation method to avoid overfitting issue even if our accuracy score was not too high.

```python
#the input size is 120 000

input_size = 120000

#choosen sizes for hidden layer
hidden_sizes = [512,256,128,64]
output_size = 6
#ouput siez = 6 because we have 6 class of defects

#We build our network with its different layers
model = nn.Sequential(nn.Linear(input_size, hidden_sizes[0]), #1st layer is linear 
                      nn.Dropout(p=0.2), #regularization method
                      nn.ReLU(), #ReLu function between layers to filter negatives files and speed up the process
                      nn.Linear(hidden_sizes[0], hidden_sizes[1]), #2nd layer is linear
                      nn.ReLU(), #ReLu function between layers to filter negatives files and speed up the process
                      nn.Linear(hidden_sizes[1], hidden_sizes[2]), #3rd layer is linear
                      nn.ReLU(), #ReLu function between layers to filter negatives files and speed up the process
                      nn.Linear(hidden_sizes[2], hidden_sizes[3]), #4th layer is linear
                      nn.ReLU(), #ReLu function between layers to filter negatives files and speed up the process
                      nn.Linear(hidden_sizes[3], output_size), #5th layer is linear 
                      nn.LogSoftmax(dim=1)) #Loss layer to compute cross entropy 
print(model)
```

```python
criterion = nn.NLLLoss()
images, labels = next(iter(train_loader))

#view(a, b)` will return a new tensor with the same data as `images` with size `(a, b)`.
images = images.view(images.shape[0], -1)

logps = model(images) #log probabilities, the output of our neural network model
loss = criterion(logps, labels) #calculate the NLL loss
```

```python
print('Before backward pass: \n', model[0].weight.grad)

loss.backward()

print('After backward pass: \n', model[0].weight.grad)

```

```python
torch.manual_seed(12)
#the Stochastic Gradient Descent optimizer
#momentum is a parameter that will avoid us getting stuck into a local minimum : a value between 0 and 1
optimizer = optim.SGD(model.parameters(), lr=0.003, momentum=0.9)
time0 = time()

#number of feedforward and backpropagation iterations that we will do to train the network
epochs = 20
for e in range(epochs):
    #we start with an error or loss of zero
    running_loss = 0
    
    #we itterate through the images dataset
    for images, labels in train_loader:
        
        # Flatten images into a long vector
        images = images.view(images.size(0),-1)
    
        # Training pass
        optimizer.zero_grad()
        
        output = model(images)
        loss = criterion(output, labels)
        
        #This is where the model learns by backpropagating
        loss.backward()
        
        #And optimizes its weights here
        optimizer.step()
        
        running_loss += loss.item()
    else:
        print("Epoch {} - Training loss: {}".format(e, running_loss/len(train_loader)))
print("\nTraining Time (in minutes) =",(time()-time0)/60)
```

```python
for images, labels in train_loader:
    print(images.shape)
```

```python
torch.manual_seed(12)
correct_count, all_count = 0, 0
for images,labels in test_loader:
  for i in range(len(labels)):
    img = images[i].view(1, 120000)
    #torch.no_grad(): Context-manager that disabled gradient calculation
    with torch.no_grad():
        logps = model(img)

    #get the scores obtained for each class. 
    ps = torch.exp(logps)
    probab = list(ps.numpy()[0])
    
    #choose the one having the maximum score and compare it to the real label
    pred_label = probab.index(max(probab))
    true_label = labels.numpy()[i]
    
    if(true_label == pred_label):
      correct_count += 1
    all_count += 1

print("Number Of Images Tested =", all_count)
print("\nNumber of images predicted correctly =", correct_count)
print("\nModel Accuracy =", (correct_count/all_count))
```

We have improved a bit our accuracy which is a good step, even if our accuracy it's still not awesome.

We have an accuracy of 58% which means that out of 180 images tested, 105 (58%) were predicted correctly
